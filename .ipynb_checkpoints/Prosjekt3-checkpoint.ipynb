```python
# HAVE TO TURN THIS INTO CLASS

HUMAN = 0
ZOMBIE = 1
N=10
nx=10
ny=10

STATE = np.repeat(HUMAN,N)
STATE[0] = ZOMBIE
Walkers = np.random.randint(0, [nx,ny],size=(N,2))

def plot_walkers():
    plt.close()
    H = Walkers[STATE == HUMAN]
    Z = Walkers[STATE == ZOMBIE]
    plt.scatter(H[:,0],H[:,1],c='b',s=120)
    plt.scatter(Z[:,0],Z[:,1],c='r',s=60)
    plt.xlim(0,nx)
    plt.ylim(0,ny)
    plt.grid()
    plt.show()

def move_walkers():
    global Walkers
    u = np.array([[1,0],[-1,0],[0,1],[0,-1]])
    dir = np.random.randint(0,4,size=N)
    Walkers=Walkers+u[dir]

def check_illegal_move():
    wrong_place_x = np.logical_or(Walkers[:,0] < 0, Walkers[:,0] > nx)
    wrong_place_y = np.logical_or(Walkers[:,1] < 0, Walkers[:,1] > ny)
    return np.logical_or(wrong_place_x, wrong_place_y)

#step 1: init
Walkers_old = np.copy(Walkers)
plot_walkers()
# check_illegal_move()
#%%
#step 2: move
for i in range(20):
    move_walkers()
    idx = check_illegal_move()
    Walkers[idx] = Walkers_old[idx]
    Walkers_old = np.copy(Walkers)
    plot_walkers()
```
